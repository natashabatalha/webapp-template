```python
import plotly.graph_objects as go
import asdf
import os
import make_plotly
import numpy as np
```

```python
## coronagraph plots here
# import asdf coronagraph + metric data

this_coronagraph = "luvoir_a_paplc"
this_metric = "angular_coverage"

newfig = go.Figure()
for this_coronagraph in ["luvoir_a_paplc", "usort_offaxis_aplc_broad"]:
    this_plot_file = os.path.join(this_coronagraph, this_metric + ".asdf")
    print(this_plot_file)
    this_file = os.path.join(os.path.abspath("./data/"), this_plot_file)
    af = asdf.open(this_file)
    
    print(af.tree)
    newfig.add_trace(go.Line(x=af.tree['r'], y=af.tree['coverage'],name=this_coronagraph))

newfig.update_layout(xaxis_title="Angular separation ($Lambda0/D$)")
newfig.update_layout(yaxis_title="Angular coverage [fraction]")
newfig.update_layout(title=f"{this_metric} metric")
newfig.update_xaxes(range=[0, 32])
newfig.update_yaxes(range=[0, 1.05])
newfig.show()

this_metric = "contrast_curve"

newfig = go.Figure()
for this_coronagraph in ["luvoir_a_paplc", "usort_offaxis_aplc_broad"]:
    this_plot_file = os.path.join(this_coronagraph, this_metric + ".asdf")
    print(this_plot_file)
    this_file = os.path.join(os.path.abspath("./data/"), this_plot_file)
    af = asdf.open(this_file)
    
    print(af.tree)
    newfig.add_trace(go.Line(x=af.tree['r'], y=af.tree['y'],name=this_coronagraph))

newfig.update_layout(yaxis_type="log")
newfig.update_layout(xaxis_title="Angular separation (\u03bb/D)")
newfig.update_layout(yaxis_title="Contrast")
newfig.update_layout(title=f"{this_metric} metric")
newfig.update_xaxes(range=[0, 32])
newfig.update_yaxes(range=[np.log10(5e-12), np.log10(2e-8)], showexponent='all', exponentformat='E')
newfig.show()

make_plotly.plot_angular_coverage(["luvoir_a_paplc", "usort_offaxis_aplc_broad"])
make_plotly.plot_contrast_curve(["luvoir_a_paplc", "usort_offaxis_aplc_broad"])

pass
```
